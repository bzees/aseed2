```r
# 필요한 패키지 설치 및 로드
install.packages("ggplot2")
library(ggplot2)

# 기본 R 명령어
x <- 1:10
y <- x^2
print(paste("x의 평균:", mean(x)))
print(paste("y의 합:", sum(y)))

# 데이터 프레임 생성
df <- data.frame(x = x, y = y)

# 기술 통계
summary(df)

# 시각화
ggplot(df, aes(x = x, y = y)) +
  geom_point() +
  geom_line() +
  labs(title = "x vs y^2", x = "x 값", y = "y 값")

# 정규분포에서 표본 추출
set.seed(123)
normal_data <- rnorm(100, mean = 0, sd = 1)

# 히스토그램
hist(normal_data, main = "정규분포 히스토그램", xlab = "값", ylab = "빈도")

# t-검정 예시
group1 <- rnorm(50, mean = 10, sd = 2)
group2 <- rnorm(50, mean = 11, sd = 2)
t_test_result <- t.test(group1, group2)
print(t_test_result)

# 상관관계 분석
correlation <- cor(x, y)
print(paste("x와 y의 상관계수:", correlation))

# 선형 회귀
lm_model <- lm(y ~ x, data = df)
summary(lm_model)

# 결과 플롯
plot(lm_model)
```
